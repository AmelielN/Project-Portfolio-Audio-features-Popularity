```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load  CSV cleaned
df_tracks = pd.read_csv("data/spotify_tracks_cleaned.csv")
```

```python
# Normalize artist names (remove ft/feat)
def normalize_text(s):
    if pd.isna(s): return s
    s = s.lower().strip()
    s = s.replace("&", " and ")
    for token in ["feat.", "ft.", "(feat", "(ft", "featuring"]:
        s = s.replace(token, "")
    return " ".join(s.split())
```
